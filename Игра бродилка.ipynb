```python
import random

def print_board(player_pos, exit_pos):
    for i in range(5):
        for j in range(5):
            if (i, j) == player_pos:
                print("P", end=" ")  # Позиция игрока
            elif (i, j) == exit_pos:
                print("E", end=" ")  # Позиция выхода
            else:
                print(".", end=" ")  # Пустая клетка
        print()

def move_player(player_pos, direction):
    # Перемещение игрока в указанном направлении
    x, y = player_pos
    if direction == "up" and x > 0:
        x -= 1
    elif direction == "down" and x < 4:
        x += 1
    elif direction == "left" and y > 0:
        y -= 1
    elif direction == "right" and y < 4:
        y += 1
    return x, y

def play_game():
    player_pos = (4, 0)  # Начальная позиция игрока
    exit_pos = (0, 4)    # Позиция выхода

    print("Добро пожаловать в игру 'Бродилка'!")
    print("Цель игры - добраться до выхода.")

    while True:
        print_board(player_pos, exit_pos)

        # Запрашиваем у игрока следующий ход
        direction = input("Выберите направление (up/down/left/right): ").lower()

        # Перемещаем игрока
        player_pos = move_player(player_pos, direction)

        # Проверяем, достиг ли игрок выхода
        if player_pos == exit_pos:
            print("Поздравляю! Вы достигли выхода. Игра завершена.")
            break

        # Добавим случайные события
        event = random.choice(["nothing", "enemy", "treasure", "trap"])
        if event == "enemy":
            print("Внимание! Вы встретили врага. Будьте осторожны!")
        elif event == "treasure":
            print("Вы нашли сокровище! Поздравляю, ваша удача.")
        elif event == "trap":
            print("О нет! Вы попали в ловушку и потеряли немного здоровья.")

if __name__ == "__main__":
    play_game()
```
